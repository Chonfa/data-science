```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import folium

# Cargar el archivo CSV
file_path = 'nuevas-estaciones-bicicletas-publicas.csv'  # Reemplaza con la ruta a tu archivo
data = pd.read_csv(file_path)

# Asegurarte de que tus datos tengan columnas de latitud y longitud
# Ejemplo:
# data.head()  # Revisa la estructura de los datos
# print(data.columns)  # Verifica los nombres de las columnas

# Crear una GeoDataFrame con los datos
geometry = [Point(xy) for xy in zip(data['LONGITUD'], data['LATITUD'])]  # Reemplaza 'longitud' y 'latitud' con los nombres de las columnas de tu archivo
gdf = gpd.GeoDataFrame(data, geometry=geometry)

# Calcular distancias al punto más cercano
gdf['distance_to_nearest'] = gdf.geometry.apply(lambda x: gdf.geometry.distance(x).min())

# Crear un mapa base centrado en Buenos Aires
mapa = folium.Map(location=[-34.6083, -58.3712], zoom_start=12)

# Agregar los puntos como un heatmap
for idx, row in gdf.iterrows():
    folium.CircleMarker(
        location=[row['LATITUD'], row['LONGITUD']],
        radius=2,
        color='blue',
        fill=True,
        fill_color='blue',
        fill_opacity=0.6,
        popup=f"Distancia: {row['distance_to_nearest']} metros"
    ).add_to(mapa)

# Mostrar el mapa
mapa
```

```python
from folium.plugins import HeatMap
mapa2 = mapa
# Crear un mapa de calor
heat_data = [[row['LATITUD'], row['LONGITUD']] for index, row in gdf.iterrows()]
HeatMap(heat_data, radius=20).add_to(mapa2)

# Guardar el mapa en un archivo HTML
mapa2
























































































































































































































```
